```python
# recommend game to the candidate
t_name = input("please enter the name of the user you intend to recommend game for:")
t_game_label = user_game['game_label'][user_game['user_name']==t_name]
t_game_label.values
t_str = str(t_game_label.values)
t_str = t_str[2:len(t_str)-2]
# t_str
t_s = set(t_str.split(','))
# t_s
res = set({})
user_s = set({})
for index, rows in user_game.iterrows():
    l5 = rows['game_label'].split(',')
    n_s = set(l5)
    if(len(t_s.intersection(n_s))>(len(t_s)/2)):
        res.add(rows['game_title'])
        user_s.add(rows['user_name'])
print('Games recommend for this user:',res)
print('Users with the same interest',user_s)
```
